```python
import pandas as pd
from tqdm import tqdm
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error as MAE
```

```python
def get_systems_at_diff_T(df):
    df['sys_key'] = df['Solute'] + '_' + df['Solvent']
    solute_solvent_sys = df['sys_key'].unique().tolist()

    ss_dict = {}
    n_points = []
    for ss in tqdm(solute_solvent_sys):
        df_spec = df[df['sys_key'] == ss]
        ss_dict[ss] = (df_spec['T'].to_numpy(), df_spec['log-omega'].to_numpy())
        n_points.append(len(df_spec))
    df_sys = pd.DataFrame({
        'Key': solute_solvent_sys,
        'n_points': n_points
    })
    return df_sys, ss_dict

def plot_system(system, ss_dict):
    Ts, ln_omegas = ss_dict[system]
    X = 1/Ts.reshape(-1,1)
    y = ln_omegas.reshape(-1,1)

    # Perform linear regression
    reg = LinearRegression().fit(X, y)
    k1 = reg.intercept_[0]
    k2 = reg.coef_[0,0]

    y_pred = reg.predict(X)
    mae = MAE(ln_omegas, y_pred)
    r2 = reg.score(X, y)

    x_line = np.linspace(min(X.reshape(-1)), max(X.reshape(-1))).reshape(-1,1)
    y_line = reg.predict(x_line)

    plt.plot(X, y, 'ro', label='Experiments')
    plt.plot(x_line, y_line, 'b--', label='Model')
    return mae, r2, k1, k2
    
```

## MN dataset

```python
df = pd.read_csv('../data/interim/MN_reps.csv')
```

```python
df_sys, ss_dict = get_systems_at_diff_T(df)
```

```python
print('Number of systems with more than or equal to 3 points: ', (df_sys['n_points']>=3).sum())
```

```python
# Get systems with at least 3 different temperatures
df_3 = df_sys[df_sys['n_points']>=3]
```

```python
with pd.option_context('display.max_rows', None, 'display.max_columns', None):  # more options can be specified also
    print(df_3.sort_values('n_points', ascending=False))
```

```python
# Example
plot_system('Poly(dimethylsiloxane)_Benzene,ethyl', ss_dict)
```

```python
systems = df_3['Key'].tolist()
K1s, K2s, maes, r2s = [], [], [], []
for system in tqdm(systems):
    # Get x and y
    Ts, ln_omegas = ss_dict[system]
    X = 1/Ts.reshape(-1,1)
    y = ln_omegas.reshape(-1,1)
    
    # Perform linear regression
    reg = LinearRegression().fit(X, y)
    
    # Store results
    K1s.append(reg.intercept_[0])
    K2s.append(reg.coef_[0,0])
    
    y_pred = reg.predict(X)
    maes.append(MAE(ln_omegas, y_pred))
    r2s.append(reg.score(X, y))

df_train_LR = df_3.copy()
df_train_LR = df_train_LR.reset_index()
df_train_LR['MAE'] = maes
df_train_LR['R2'] = r2s
df_train_LR['K1'] = K1s
df_train_LR['K2'] = K2s
```

```python
print('MAE: ', np.round(np.mean(maes),2), '+-', np.round(np.std(maes),3))
print('R2: ', np.round(np.mean(r2s),2), '+-', np.round(np.std(r2s),3))
```

```python
threshold = 0.90
percent = (df_train_LR['R2']>=threshold).sum()/len(df_train_LR)*100
print('Percentage of systems with R2 above '+str(threshold)+': ', np.round(percent,2))
```

```python
df_train_LR = df_train_LR.drop(columns=['index'])
df_train_LR.to_csv('../data/interim/MN_LR.csv', index=False)
```

```python
df_worse = df_train_LR.sort_values('R2')
df_worse
```

```python
system = 'Poly(n-butyl methacrylate)_Benzene,tert-butyl'
_ = plot_system(system, ss_dict)
```
