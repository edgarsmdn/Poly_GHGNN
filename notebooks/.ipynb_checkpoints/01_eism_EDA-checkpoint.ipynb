```python
import pandas as pd
import matplotlib as mpl
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
```

```python
import os

folder = '../reports/figures/distributions'
if not os.path.exists(folder):
    os.makedirs(folder)
```

## MN dataset

```python
df = pd.read_csv('../data/interim/MN_reps.csv')
```

```python
fig = sns.displot(df, kde=True, x='ln-omega', legend=True, height=6, aspect=2)
plt.xlabel('ln($\Omega^\infty$)', fontsize=18)
#plt.xlim(-2,10)
plt.ylabel('Count', fontsize=18)
plt.title('MN data set', fontsize=18)
fig.savefig(folder + '/MN_omega.png', dpi=350)
plt.show()
```

```python
fig = sns.displot(df, kde=True, x='ln-MN', legend=True, height=6, aspect=2)
plt.xlabel('ln(MN)', fontsize=18)
#plt.xlim(4,16)
plt.ylabel('Count', fontsize=18)
plt.title('MN data set', fontsize=18)
fig.savefig(folder + '/MN_MN.png', dpi=350)
plt.show()
```

```python
fig = sns.displot(df, kde=True, x='T', legend=True, height=6, aspect=2)
plt.xlabel('Temperature (K)', fontsize=18)
#plt.xlim(250, 600)
plt.ylabel('Count', fontsize=18)
plt.title('MN data set', fontsize=18)
fig.savefig(folder + '/MN_T.png', dpi=350)
plt.show()
```
